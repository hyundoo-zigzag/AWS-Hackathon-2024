```python
import boto3
import json
from IPython.display import Markdown
```

```python
bedrock = boto3.client('bedrock-runtime', region_name='ap-northeast-2')
```

```python
system_prompt = '''
[role]
- 당신은 주어진 회의 기록을 정리하여 마크다운 문서로 정리합니다.
- 회의 내용 중 해야 할 일이 있다면 어떤 사람이 어떤 일을 해야 하는지도 같이 정리해주세요.

[current_year]
2024

[additional_informations]
- 회의는 화상으로 진행되었습니다.
- 해야 할 일은 불릿포인트로 정리해주세요.

[response_format]
- 응답은 정리된 마크다운 문서만 답변합니다.
- 마크다운 문서 외에는 답변하지 마세요.
'''
```

```python
with open('asrOutput.json', 'r') as f:
    meeting_script = json.load(f)
```

```python
scripts = ''
speaker = meeting_script['results']['audio_segments'][0]['speaker_label']
for script in meeting_script['results']['audio_segments']:
    if speaker != script['speaker_label']:
        scripts += '\n'+script['speaker_label'] + ' : '
        speaker = script['speaker_label']
    scripts += script['transcript'] + ' '
```

```python
print(scripts)
```

```python
model_id = 'anthropic.claude-3-5-sonnet-20240620-v1:0'
# model_id = 'anthropic.claude-3-haiku-20240307-v1:0'
```

```python
messages =  [{"role": "user", "content": scripts}]
```

```python
body = json.dumps(
    {
        "anthropic_version": "bedrock-2023-05-31",
        "max_tokens": 8192,
        "system": system_prompt,
        "messages": messages
    }  
)
```

```python
response = bedrock.invoke_model(body=body, modelId=model_id)
```

```python
response_body = json.loads(response.get('body').read())
```

```python
Markdown(response_body['content'][0]['text'])
```
